```python
x = np.linspace(-4,4,1000)
```

```python
tanh_data = np.tanh(x)
```

```python
trace0 = go.Scatter(
    x=x,
    y=tanh_data
)

plotly.offline.iplot([trace0])
```

```python
# OLS model

model = sm.OLS(tanh_data,x).fit()
```

```python
ols_predict = model.predict()
```

```python
trace0 = go.Scatter(
    x=x,
    y=tanh_data
)

trace1 = go.Scatter(
    x=x,
    y=ols_predict
)

plotly.offline.iplot([trace0,trace1])
```
